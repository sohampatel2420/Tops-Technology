Essay: Importance of Data Cleaning in Data Science
Introduction
Data cleaning, also known as data cleansing or data preprocessing, is a critical step in the data science workflow. It involves detecting and correcting (or removing) corrupt, inaccurate, incomplete, or irrelevant parts of the data. Clean data leads to better model performance, accurate insights, and trustworthy results.

Why Data Cleaning Is Important

Improves Data Quality: Raw data often contains missing values, outliers, or errors. Cleaning ensures the data is consistent and accurate.

Enhances Model Performance: Machine learning algorithms rely on high-quality data. Dirty data can reduce model accuracy or introduce bias.

Reduces Noise and Redundancy: Cleaning eliminates unnecessary variables or duplicates, helping models focus on meaningful features.

Prevents Misleading Insights: Clean data ensures correct conclusions, reducing the risk of making decisions based on flawed analysis.

Saves Time in the Long Run: Investing time in data cleaning early prevents downstream issues during analysis and modeling.

Common Data Cleaning Tasks

Handling missing values (e.g., imputation, deletion)

Removing duplicates

Correcting inconsistent formats

Filtering out outliers

Normalizing or standardizing data

Conclusion
Data cleaning is often seen as tedious, but it is one of the most essential tasks in data science. Without it, even the most advanced algorithms or sophisticated models will not yield reliable results. Clean data is the foundation of good data science.

In this column some values are simmiler so that i will combine that values: 


reguler unleaded ------> petrol

premium unleaded (reqired) ------> petrol

reguler unleaded (recomended) ------> petrol

flex-fuel (unleaded/E85) ------> flex-fuel

--> Reduces noise in the data and helps with faster and more stable training.

```python
change_name = {
    'regular unleaded': 'petrol',
    'premium unleaded (required)': 'petrol',
    'premium unleaded (recommended)': 'petrol',
    'flex-fuel (unleaded/E85)': 'flex-fuel',
    'flex-fuel (premium unleaded required/E85)': 'flex-fuel',
    'flex-fuel (premium unleaded recommended/E85)': 'flex-fuel',
    'diesel': 'diesel',
    'electric': 'electric',
    'natural gas': 'gas'
}

df['Engine Fuel Type'] = df['Engine Fuel Type'].map(change_name)
```

## DATA VISUALIZATION

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_makes = df['Make'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_makes.index, y=top_makes.values)
plt.title('Top 10 Vehicle Makes')
plt.ylabel('Count')
plt.xlabel('Make')
plt.show()
```

### Insights :

- Chevrolet is the most common vehicle make, with a significantly higher count than all others.

- Mazda, Cadillac, and GMC are the least common among the top 10, showing similar and lower frequencies.

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Engine HP'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Engine Horsepower')
plt.xlabel('Engine HP')
plt.ylabel('Frequency')
plt.show()
```

### Insights :

- Most vehicles have engine horsepower between 150 and 300, that is common and most frequent.

- Very few vehicles have horsepower above 600, showing a right-skewed distribution with rare high-power engines.

```python
plt.figure(figsize=(18,6))
style_hp = df.groupby('Vehicle Style')['Engine HP'].mean().sort_values(ascending=False)
sns.barplot(x=style_hp.index, y=style_hp.values)
plt.title('Average Engine HP by Vehicle Style')
plt.ylabel('Average Engine HP')
plt.xlabel('Vehicle Style')
plt.xticks(rotation=45)
plt.show()
```

### Insights :

- Coupes and Convertibles have the highest average engine horsepower, suggesting a focus on performance.

- Hatchbacks and Convertible SUVs have the lowest average horsepower, indicating they are more suited for daily or utility use.

- 4 Door SUVs and Sedans are best velue for money choices car styles.

```python

sns.boxplot(data=df, x='Transmission Type', y='Engine HP')
plt.title('Engine HP by Transmission Type')
plt.show()
```

### Insights :

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Engine Cylinders', y='Engine HP', data=df)
plt.title('Engine HP by Number of Cylinders')
plt.xlabel('Engine Cylinders')
plt.ylabel('Engine Horsepower')
plt.show()
```

### Insights : 

- Engine horsepower generally increases with the number of cylinders, indicating a strong positive correlation.

- 6-cylinder and 8-cylinder engines have the widest variability in horsepower, with many outliers, suggesting a range of performance levels.

- 10, 12, and 16-cylinder engines have consistently high horsepower, with fewer extreme outliers, reflecting their performance-oriented nature.

- 3 and 4-cylinder engines are associated with the lowest horsepower and are likely used in economy or compact vehicles.
